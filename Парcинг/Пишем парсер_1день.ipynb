```python
file = open('for_parsing.html', 'r', encoding ='utf-8')
```

```python
data = file.read()
file.close()
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'lxml')
```

```python
# возвращается самый первый тег(не все)
print(soup.body)
```

```python
print(soup.h2.name)
print(soup.h2.text)
```

```python
for tag in soup.a.recursiveChildGenerator():
    print(tag)
```

```python
reviews = soup.findAll('div', attrs ={'class':'review'})
for review in reviews:
    name=review.find('div', attrs ={'class':'name'}).text
    comment=review.find('div', attrs ={'class':'comment'}).text
    print(name.strip('').strip('/n'), comment, sep = ':')
```

```python
for i in soup.findAll('li'):
    print(str(i).replace('<li>', '').replace('</li>', ''))
```

```python
file = open('market.html', 'r', encoding ='utf-8')
data = file.read()
file.close()
```

```python
soup = BeautifulSoup(data, 'lxml')
# возвращается самый первый тег(не все)
values = soup.findAll('div', attrs ={'itemprop':'review'})
```

```python
box =[]
for value in values:
    author_tag = value.find('meta', attrs ={'itemprop':'author'})['content']
    description_tag = value.find('meta', attrs ={'itemprop':'description'})['content'].split()
   
    b = []
    for i in ['Опыт использования:', 'Достоинства:', 'Недостатки:', 'Комментарий:']:
        try:
            a = description_tag.index(i)
            b.append(a)
        except:
            pass

    for i in range(len(b)):
        if i == len(b)-1:
            a1, b1 = description_tag[b[i]:][0].replace(':', ''), ' '.join(description_tag[b[i]:][1:])
            box.append([author_tag, a1, b1])

        else:
            a1, b1 = description_tag[b[i]: b[i+1]][0].replace(':', ''),' '.join(description_tag[b[i]: b[i+1]][1:])
#             box.append([author_tag, ' '.join(description_tag[b[i]: b[i+1]])])
            box.append([author_tag, a1, b1])
```

```python
import pandas as pd
pd.DataFrame(box)
```
