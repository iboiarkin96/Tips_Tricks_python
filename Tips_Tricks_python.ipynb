# Как делать подписи на графиках 

```python
listPepper = [ 
    [5, "Bell pepper", "Not even spicy"], 
    [6, "Espelette pepper", "Uncomfortable"], 
    [20, "Chocolate habanero", "Practically ate pepper spray"]]
_x = pd.DataFrame(listPepper, columns = ['count', 'first', 'col'])
```

```python

xs = _x['first']
ys = _x['count']

plt.bar(xs,ys)

for x,y in zip(xs,ys):

    label = "{:.2f} ".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.title('title')
plt.show()
```
